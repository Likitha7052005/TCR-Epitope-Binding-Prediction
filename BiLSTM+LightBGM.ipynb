```python
# LightGBM + BiLSTM (FULLY FIXED VERSION)
# ============================================

import pandas as pd
import numpy as np
import random
import lightgbm as lgb
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.layers import Input, Embedding, LSTM, Bidirectional, Dense, Concatenate, Dropout
from tensorflow.keras.models import Model

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)

# ============================================
# 1. LOAD DATASET
# ============================================

file_path = "McPAS-TCR.csv"
df = pd.read_csv(file_path, low_memory=False)

# Detect important columns
tcr_col = [c for c in df.columns if "CDR3.beta" in c][0]
epi_col = [c for c in df.columns if "Epitope" in c or "epitope" in c.lower()][0]

df = df[[tcr_col, epi_col]].dropna()
df.columns = ["tcr", "epi"]

# Clean short sequences
df = df[df["tcr"].str.len() >= 6]
df = df.drop_duplicates().reset_index(drop=True)

df["binding"] = 1  # all positive
print("Positive samples:", len(df))


# ============================================
# 2. BETTER NEGATIVE SAMPLING
# ============================================

def generate_negatives(pos_df, n_neg_per_pos=3):
    tcrs = pos_df["tcr"].values
    epis = pos_df["epi"].values
    pos_set = set(zip(tcrs, epis))
    negs = []
    tries = 0
    target = n_neg_per_pos * len(pos_df)

    while len(negs) < target and tries < target * 10:
        tries += 1
        t = np.random.choice(tcrs)
        e = np.random.choice(epis)
        if (t, e) not in pos_set:
            negs.append((t, e, 0))

    neg_df = pd.DataFrame(negs, columns=["tcr", "epi", "binding"])
    return neg_df.drop_duplicates().reset_index(drop=True)

neg_df = generate_negatives(df, n_neg_per_pos=3)
print("Negative samples:", len(neg_df))

full = pd.concat([df, neg_df], ignore_index=True)
full = full.sample(frac=1, random_state=RANDOM_SEED).reset_index(drop=True)

print("Full dataset:", len(full))
print("Positive ratio:", full["binding"].mean())


# ============================================
# 3. CREATE K-MER TF-IDF FEATURES (LightGBM)
# ============================================

def kmers(seq, k=3):
    seq = seq.strip()
    return " ".join([seq[i:i+k] for i in range(len(seq) - k + 1)])

full["tcr_k"] = full["tcr"].apply(lambda x: kmers(x, 3))
full["epi_k"] = full["epi"].apply(lambda x: kmers(x, 3))

tf_tcr = TfidfVectorizer(min_df=2)
tf_epi = TfidfVectorizer(min_df=2)

X_tcr = tf_tcr.fit_transform(full["tcr_k"])
X_epi = tf_epi.fit_transform(full["epi_k"])

from scipy.sparse import hstack
X = hstack([X_tcr, X_epi]).tocsr()
y = full["binding"].values


# ============================================
# 4. LIGHTGBM MODEL (FIXED FOR OLD VERSIONS)
# ============================================

print("\n🔵 Training LightGBM with 5-fold CV...")

kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_SEED)
fold_auc = []

params = {
    "objective": "binary",
    "metric": "auc",
    "learning_rate": 0.05,
    "num_leaves": 31,
    "seed": RANDOM_SEED,
    "verbose": -1
}

fold = 1
for train_idx, val_idx in kfold.split(X, y):
    print(f"Fold {fold}...")
    Xtr, Xval = X[train_idx], X[val_idx]
    ytr, yval = y[train_idx], y[val_idx]

    train_data = lgb.Dataset(Xtr, label=ytr)
    val_data = lgb.Dataset(Xval, label=yval)

    model = lgb.train(
        params,
        train_data,
        num_boost_round=1000,
        valid_sets=[val_data],
        valid_names=["valid"],
        callbacks=[lgb.early_stopping(50)]   # works on all versions
    )

    preds = model.predict(Xval)
    auc = roc_auc_score(yval, preds)
    fold_auc.append(auc)
    print("Fold AUC:", auc)
    fold += 1

print("\nMean AUC:", np.mean(fold_auc))


# ============================================
# 5. FINAL TEST SPLIT FOR LightGBM
# ============================================

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=RANDOM_SEED
)

train_data = lgb.Dataset(X_train, label=y_train)
test_data = lgb.Dataset(X_test, label=y_test)

final_model = lgb.train(
    params,
    train_data,
    num_boost_round=1000,
    valid_sets=[test_data],
    valid_names=["test"],
    callbacks=[lgb.early_stopping(50)]
)

preds = final_model.predict(X_test)
pred_labels = (preds >= 0.5).astype(int)

print("\n🎯 LightGBM Test Results:")
print("Accuracy:", accuracy_score(y_test, pred_labels))
print("AUC:", roc_auc_score(y_test, preds))
print(classification_report(y_test, pred_labels))

cm = confusion_matrix(y_test, pred_labels)
sns.heatmap(cm, annot=True, fmt='d')
plt.title("LightGBM Confusion Matrix")
plt.show()


# ============================================
# 6. DEEP LEARNING MODEL (BiLSTM)
# ============================================

print("\n🔵 Training BiLSTM model...")

# Character encoding
chars = sorted(list(set("".join(full["tcr"]) + "".join(full["epi"]))))
char_to_id = {c: i+1 for i, c in enumerate(chars)}

def encode(seq):
    return [char_to_id[c] for c in seq]

tcr_ids = [encode(s) for s in full["tcr"]]
epi_ids = [encode(s) for s in full["epi"]]

from tensorflow.keras.preprocessing.sequence import pad_sequences
max_len = max(max(len(x) for x in tcr_ids), max(len(x) for x in epi_ids))

tcr_pad = pad_sequences(tcr_ids, maxlen=max_len, padding='post')
epi_pad = pad_sequences(epi_ids, maxlen=max_len, padding='post')

X_tcr_train, X_tcr_test, X_epi_train, X_epi_test, y_train2, y_test2 = train_test_split(
    tcr_pad, epi_pad, y, test_size=0.2, stratify=y, random_state=RANDOM_SEED
)

vocab_size = len(chars) + 1

# Build BiLSTM
inp_tcr = Input(shape=(max_len,))
inp_epi = Input(shape=(max_len,))

emb_tcr = Embedding(vocab_size, 32, mask_zero=True)(inp_tcr)
emb_epi = Embedding(vocab_size, 32, mask_zero=True)(inp_epi)

lstm_tcr = Bidirectional(LSTM(64))(emb_tcr)
lstm_epi = Bidirectional(LSTM(64))(emb_epi)

x = Concatenate()([lstm_tcr, lstm_epi])
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)
out = Dense(1, activation='sigmoid')(x)

seq_model = Model([inp_tcr, inp_epi], out)
seq_model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

seq_model.summary()

seq_model.fit(
    [X_tcr_train, X_epi_train],
    y_train2,
    validation_split=0.15,
    epochs=12,
    batch_size=64,
    verbose=1
)

pred2 = (seq_model.predict([X_tcr_test, X_epi_test]) >= 0.5).astype(int)

print("\n🎯 BiLSTM Results:")
print("Accuracy:", accuracy_score(y_test2, pred2))
print(classification_report(y_test2, pred2))

cm2 = confusion_matrix(y_test2, pred2)
sns.heatmap(cm2, annot=True, fmt='d')
plt.title("BiLSTM Confusion Matrix")
plt.show()
```
